```python
import matplotlib.pyplot as plt
import numpy as np
def scatter_heat(list):
    '''
    input a list of tuples, output a scatter plot with heat map
    '''
    arr=np.array(list)
    plt.scatter(arr[:,0],arr[:,1],c=arr[:,2],cmap='viridis')
    plt.xlabel('critic')
    plt.ylabel('actor')
    plt.colorbar()
    plt.show()
scatter_heat(list1)
```
